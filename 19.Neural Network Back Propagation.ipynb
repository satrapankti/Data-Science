```python
import numpy as np
import matplotlib.pyplot as plt
from keras.models import Sequential # generate networks/layers
from keras.layers import Dense # creates layer
```

```python
# fix random seed for reproducibility
seed = 7
np.random.seed(seed)

dataset=np.loadtxt("/11.pima-indians-diabetes.csv"
                      ,delimiter=",")
# split data into X and Y variables
X = dataset[:,0:8]
Y = dataset[:,8]
```

```python
# create model/adding layer to model
model = Sequential()
model.add(Dense(12, input_dim=8, activation="relu"))  #1st hidden layer,input layer
model.add(Dense(8, kernel_initializer="uniform", activation="relu")) #2nd hidden layer
model.add(Dense(1, kernel_initializer="uniform", activation="sigmoid")) #output layer
```

```python
model.summary()
```

```python
# Compile model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
#loss=backpropogation calculation method,optimizer=GD variants,metrics=classification table in logistic regression-o/p is binary
```

```python
# Fit the model
story=model.fit(X, Y, validation_split=0.33, epochs=300, batch_size=10)
#train->loss-accuracy,test->loss-accuracy
```

```python
story.params
```

**%** is a place holder and **f** is for float,**s** is for string.
This is old print statement formatting.
print("**% _** " **%** ( **_** )) -> print("**{}**"**.format**( **_** ))
,**%s** it will print for string;
**%f** it will print for float, **%.f** will round off the value to an integer, 
**%.2f**->(2) will work for no. of decimals required by default **%f** will print **6** rounded decimals, 
**%f%%** the last **%%** after f is for printing the %percentage sign.

```python
# evaluate the model and print its accuracy 
scores = model.evaluate(X, Y)
print("%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
```

```python
# list all data in history from story as dictionary in key format from keras 
story.history.keys()
```

```python
# visualize history for accuracy & loss
fig = plt.figure(figsize=(17,5))
plt.suptitle("Model Accuracy Vs Loss", size=20, y=1.07) 
plt.subplot(1,2,1)
plt.plot(story.history['accuracy'],color="gold")
plt.plot(story.history['val_accuracy'],color="c")
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper left')

plt.subplot(1,2,2)
plt.plot(story.history['loss'],color="gold")
plt.plot(story.history['val_loss'],color="c")
plt.title('Model Loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Test'], loc='upper right')
plt.show()
```
