## Seaborn

```python
tips =sns.load_dataset('tips')
```

```python
tips.head()
```

```python
sns.get_dataset_names()  #to find the data sets inbuild in seaborn
```

### Univariate  one variable data visualization

```python
#Strip Plot
#it is a single axis-scatter plot
sns.stripplot(y= 'tip', data=tips,jitter=True,Color="teal")  #y=column,data=df,jitter=T/F,jitter is to spread out 
plt.ylabel('tip ($)') 
plt.show()
```

```python
#Grouping with stripplot()
#for more than 1 column
sns.stripplot(x='day', y='tip', data=tips,size=4,jitter=False) #x=column1 &y=column 2,size=size of points
plt.ylabel('tip ($)')
plt.show()
```

```python
#Swarm plot
#the more it is spread out the max is the range.same as strip plot but spread out
#orient is for h-horiz,y-vert
sns.swarmplot(x='day', y='tip', data=tips,size=4,orient="v")
plt.ylabel('tip ($)')
plt.show()
```

```python
#based on parameters
sns.swarmplot(x='tip', y='day', data=tips, hue='sex',orient='h',size=3)   # hue -distinguish based on  parameters
plt.xlabel('tip ($)')
plt.show()
```

```python
#Box Plot
sns.boxplot(x='day', y='tip', data=tips)  # can also add hue to distinguish
plt.ylabel('tip ($)')
plt.show()
```

```python
#Violin Plot
sns.violinplot(x='day', y='tip', data=tips) # can also add hue to distinguish
plt.ylabel('tip ($)')
plt.show()
```

```python
#adding Subplots
plt.subplot(1,2,1)   # (r,c,1st plot)
sns.boxplot(x='day', y='tip', data=tips)    # (r,c,2nd plot)
plt.ylabel('tip ($)')
plt.subplot(1,2,2)
sns.violinplot(x='day', y='tip', data=tips)
plt.ylabel('tip ($)')

plt.show()
```

```python
#Combining plots

sns.violinplot(x='day', y='tip', data=tips, inner=None,color='lightgray')
sns.stripplot(x='day', y='tip', data=tips, size=4,jitter=True)
plt.ylabel('tip ($)')
plt.show()
```

### Bivariate  two variables data visualization

```python
#Joint plot
sns.jointplot(x= 'total_bill', y= 'tip', data=tips)  # can also add hue to distinguish
plt.show()
```

```python
# Density plot
sns.jointplot(x='total_bill', y= 'tip', data=tips,kind='kde')  #kind=kde-for density plot
plt.show()
```

```python
#Pair plot
# it compares all the variables with each other in df
sns.pairplot(tips,hue="sex")
plt.show()
```
