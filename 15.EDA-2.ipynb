## 4.Isolation Forest

```python
data_encoded=pd.get_dummies(iris)
data_encoded.tail()
```

```python
# training the model
clf = IsolationForest(random_state=10,contamination=.05)
clf.fit(data_encoded)
```

```python
# predictions
y_pred_outliers = clf.predict(data_encoded)
y_pred_outliers  #-1 for outliers and 1 for inliers.
```

```python
data_encoded
```

```python
# Let us add a new data point which is outlier
data_encoded.loc[150]=[20,40,30,50,1,0,0]
```

```python
data_encoded
```

```python
# again train the model and predict
clf = IsolationForest(random_state=10,contamination=.01)
clf.fit(data_encoded)
y_pred_outliers = clf.predict(data_encoded)
y_pred_outliers
```

```python
# creating a column for anomoaly scores and outliers and add to original df
data_encoded['scores']=clf.decision_function(data_encoded)
```

```python
data_encoded['anomaly']=clf.predict(data_encoded.iloc[:,0:7])
```

```python
data_encoded
```

```python
#Print the outlier data points
data_encoded[data_encoded['anomaly']==-1]
```
