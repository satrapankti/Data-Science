```python
import pandas as pd
from pandas import DataFrame
import matplotlib.pyplot as plt
from pandas import Grouper
from pandas.plotting import lag_plot
from statsmodels.graphics.tsaplots import plot_acf
```

**Line Plot**

```python
series = pd.read_csv('D:/ExcelR/Data Science(Andheri-Sadanand-Chirag)(21st Aug 2021)/Data Set Files/24.daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True)
series.head()
```

```python
plt.suptitle('Line Plot', size=15, y=0.94)
series.plot(color="seagreen",figsize=(16,4))
plt.show()
```

**Histogram**

```python
series.hist(grid=False,facecolor ="yellow",edgecolor ="black",figsize=(10,4))
plt.show()
```

**Density Plot**

```python
series.plot(kind='kde',color="indigo",figsize=(16,4))
plt.suptitle('Density Plot', size=15, y=0.94)
plt.show()
```

**Box and Whisker Plots by Interval**

```python
series = pd.read_csv('D:/ExcelR/Data Science(Andheri-Sadanand-Chirag)(21st Aug 2021)/Data Set Files/24.daily-minimum-temperatures.csv', header=0, index_col=0,parse_dates=True,squeeze=True)
groups = series.groupby(Grouper(freq='A'))
years = DataFrame()
for name, group in groups:
    years[name.year] = group.values
years.boxplot(figsize=(16,4))
plt.suptitle('Box Plot', size=15, y=0.94)
plt.show()
```

**Lag plot**

```python
fig = plt.figure(figsize=(10,5))
plt.suptitle('Lag Plot', size=15, y=0.93) 
lag_plot(series)
plt.show()
```

 **Autocorrelation Plot**

```python
fig, ax = plt.subplots(figsize=(16,4))
plot_acf(series,lags=100,title="ACF Plot",color="r",ax=ax)
plt.show()
```
